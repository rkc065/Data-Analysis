Most number of 4's scored in a Season was year 2013.In Year 2013 almost closed to 2300 forus was scored

Total number of 6's Scored in each of the season

```python
sixs_list = []
for var in data1:
    df_new = matches_df[matches_df['Season']==var]
    total_sixes = 0
    for i in df_new['id'].values:
        temp = ball_by_ball[ball_by_ball['id']==i]
        sixes= temp[temp['batsman_runs']==6]['batsman_runs'].count()
        total_sixes+=sixes
    sixs_list.append(total_sixes)
    
ax = plt.axes()
ax.set(facecolor = "black")   
sns.barplot(x=data1, y=sixs_list,saturation=1)
#plt.plot(data1,fours_list,'s-r')
plt.title('Total number of Sixes scored in each season',fontsize=15,fontweight="bold")
plt.xlabel('\n Season',fontsize=10)
plt.ylabel('\n Total Fours',fontsize=10)
plt.show()
```

From the Above analysis we can say that most number of sixes scored in a season was year 2018 wih almost close to 900 sixes
