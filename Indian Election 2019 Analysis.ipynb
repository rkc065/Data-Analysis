#  Q.) How many Constituency is there in each of the indian state?

```python
Cons_per_state=election_data.groupby(['State'])['Constituency'].count().reset_index().rename(columns={'Constituency':'Constituency'})
```

```python
Cons_per_state
```

```python
color_list=['black', 'red', 'green', 'blue', 'pink','orange','coral']
```

```python
plt.figure(figsize=(15,10))
plt.bar(Cons_per_state['State'], Cons_per_state['Constituency'],color=color_list);
#plt.plot(Cons_per_state['State'], Cons_per_state['Constituency'],'p--g');
plt.xlabel('State Name')
plt.xticks(rotation=90,fontsize=10)
plt.ylabel('No of Constituency')
plt.title('Number of Constituencys in each state');
```

Which Indian state is having maximum number of loksabha Constituency?

```python
max_const=Cons_per_state[Cons_per_state.Constituency==Cons_per_state.Constituency.max()].reset_index().rename(columns={'Constituency':'Count'})
```

```python
max_const
```

# How many parties won atleast one seat in the election?

```python
#parties_list=election_data.groupby('Party').count()
parties=election_data.groupby(['Party'])['Constituency'].count().reset_index()
```

```python
len(parties)
```

Name of the parties of have won alleast one seat in the election

```python
print(parties['Party'])
```

```python
parties_won=election_data.groupby(['Party'])['Constituency'].count().reset_index().rename(columns={'Constituency':'No of set won'})
```

```python
parties_won.sort_values(by=['No of set won'], inplace=True,ascending=False)
```

```python
parties_won
```

```python
plt.figure(figsize=(20,15))
plt.bar(parties_won['Party'],parties_won['No of set won'],color=color_list)
plt.xlabel('Party Name')
plt.xticks(rotation=90,fontsize=15)
plt.ylabel('No of set won')
plt.title('Number of seat won by each party');
```

```python
ax = plt.axes()
ax.set(facecolor = "black")
sns.set(rc={'figure.figsize':(15,10)},style='whitegrid')
ax.set_title('Number of Seat won by each party ',fontsize=15,fontweight="bold")
sns.barplot(y=parties_won['Party'], x=parties_won['No of set won'], orient='h',saturation=1)
plt.xlabel('No. of Seat won')
plt.ylabel('Party')
plt.show()
```
